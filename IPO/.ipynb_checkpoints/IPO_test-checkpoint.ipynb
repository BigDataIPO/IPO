```python
Traindata = ipo_kq.drop(['종목코드', '수요예측일\r\n(시작)', '수요예측일\r\n(마감)', '청약일\r\n(시작)',
       '청약일\r\n(마감)', '납입일', '희망공모가격(원)',  '공모주식수(주)',
       '우리사주\r\n배정물량(주)', '기관투자자\r\n배정물량(주)', '일반투자자\r\n배정물량(주)', '상장(예정)일',
       '공모가(원).1','공모금액(천원)', '공모주식수(주).1', '신주모집(주)', '구주매출(주)',
       '경쟁률', '상장주식수(주)', '유통가능주식수(주)', '상장주선인'], axis = 1)
```

```python
Traindata['상장일 수정주가'] = np.nan
```

```python
Traindata
```

```python
IPOcoms = Traindata.index
```

```python
IPOcoms
```

```python
for i in IPOcoms:
    IPOday = Traindata.loc[i,'상장일']
    Adj_price = trading_kq.loc[(i,'수정주가'),IPOday]
    Traindata.loc[i,'상장일 수정주가'] = Adj_price[0]
```

```python
Traindata
```

```python
def MatchItem(IPOcoms, FeatureDf, TargetDf, indexer, finding):
    """
    IPO상장기업에 대해서 특정날(indexer)에 찾고자하는 지표(finding)을 TargetDf에서 찾은뒤 FeatureDf에 indexer날에 넣어줌
    """
    FeatureDf[indexer + finding] = np.nan
    for i in IPOcoms:
        IPOday = FeatureDf.loc[i,indexer]
        Adj_price = TargetDf.loc[(i,finding),IPOday]
        FeatureDf.loc[i,indexer + finding] = Adj_price[0]
    return FeatureDf
```

```python
Traindata = MatchItem(IPOcoms, Traindata, trading_kq, '상장일', '수정주가')
```
